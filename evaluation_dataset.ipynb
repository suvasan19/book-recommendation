```python
def jaccard_similarity(list1, list2):
    #print("doc1", doc1[0], "doc2", doc2[0])
   # union = len(set(doc1).union(set(doc2)))
    #intersection = len(set(doc1).intersection(set(doc2)))
    #print("union ", union)
    #print("intersection ", intersection)

    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection)/union
```

```python
def ranked_jaccard_similarity(index):
    ranks = []
    reviews = df['review_sentences'][index]
    #print(reviews)
    for i,row in df.iterrows():
        if i == index:
            continue
        else:
            rev = row['review_sentences']
     #       print(rev)
            js = jaccard_similarity(reviews, rev)
            # if(js > 0.0001):
            #   # print("jaccard", js)
            ranks.append((row.title, js))

    return sorted(ranks, key=lambda x: x[1], reverse = True)
```

```python
print(ranked_jaccard_similarity(139)[0:10])
```

```python
eval_rankings = {}

for book in eval:
  eval_rankings[df["title"][book]] = ranked_jaccard_similarity(book)[0:10]
```
